```python
import pandas as pd

# Load the training and testing datasets
train_df = pd.read_csv('AdFlush_train.csv')
test_df = pd.read_csv('AdFlush_test.csv', engine='python', on_bad_lines='skip')

print("Training Data:")
display(train_df.head())

print("\nTesting Data:")
display(test_df.head())
```

```python
# Check for missing values in the training and testing datasets
print("Missing values in training data:")
display(train_df.isnull().sum())

print("\nMissing values in testing data:")
display(test_df.isnull().sum())
```

```python
# Impute missing values in the testing data with the mean
for column in test_df.select_dtypes(include='number').columns:
    if test_df[column].isnull().any():
        mean_value = test_df[column].mean()
        test_df[column].fillna(mean_value, inplace=True)

print("Missing values in testing data after imputation:")
display(test_df.isnull().sum())
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# Separate features (X) and target (y)
X_train = train_df.drop('label', axis=1)
y_train = train_df['label'].astype(int)
X_test = test_df.drop('label', axis=1)
y_test = test_df['label'].astype(int)

# Convert 'ng_15_0_15' column to numeric, coercing errors and impute any resulting NaNs
X_test['ng_15_0_15'] = pd.to_numeric(X_test['ng_15_0_15'], errors='coerce')
if X_test['ng_15_0_15'].isnull().any():
    mean_value = X_test['ng_15_0_15'].mean()
    X_test['ng_15_0_15'].fillna(mean_value, inplace=True)


xgb_model = xgb.XGBClassifier(objective='binary:logistic', random_state=42)
xgb_model.fit(X_train, y_train)

y_pred = xgb_model.predict(X_test)
y_pred_proba = xgb_model.predict_proba(X_test)[:, 1]

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_pred_proba)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print(f"AUC: {auc:.4f}")
```

```python
from sklearn.model_selection import GridSearchCV
import numpy as np

# smaller grid of hyperparameters
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [3, 5],
    'learning_rate': [0.1, 0.01]
}

# Using a subset of the training data for faster tuning
X_train_sample, _, y_train_sample, _ = train_test_split(X_train, y_train, train_size=0.3, random_state=42, stratify=y_train)

grid_search = GridSearchCV(estimator=xgb.XGBClassifier(objective='binary:logistic', random_state=42),
                           param_grid=param_grid,
                           scoring='roc_auc', # Using AUC as the scoring metric
                           cv=3, # Using 3-fold cross-validation
                           verbose=2,
                           n_jobs=-1)

grid_search.fit(X_train_sample, y_train_sample)

print(f"Best hyperparameters: {grid_search.best_params_}")
print(f"Best AUC score from cross-validation: {grid_search.best_score_:.4f}")

best_xgb_model = xgb.XGBClassifier(**grid_search.best_params_, objective='binary:logistic', random_state=42)
best_xgb_model.fit(X_train, y_train)

y_pred_tuned = best_xgb_model.predict(X_test)
y_pred_proba_tuned = best_xgb_model.predict_proba(X_test)[:, 1]

accuracy_tuned = accuracy_score(y_test, y_pred_tuned)
precision_tuned = precision_score(y_test, y_pred_tuned)
recall_tuned = recall_score(y_test, y_pred_tuned)
f1_tuned = f1_score(y_test, y_pred_tuned)
auc_tuned = roc_auc_score(y_test, y_pred_proba_tuned)

print("\nTuned Model Performance on Test Set:")
print(f"Accuracy: {accuracy_tuned:.4f}")
print(f"Precision: {precision_tuned:.4f}")
print(f"Recall: {recall_tuned:.4f}")
print(f"F1 Score: {f1_tuned:.4f}")
print(f"AUC: {auc_tuned:.4f}")
```

```python
import xgboost as xgb
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold
from scipy.stats import uniform, randint
import numpy as np
import time

print("\n[1/3] Running Randomized Search...")
print("-" * 70)

param_distributions = {
    # Number of trees
    'n_estimators': randint(100, 1000),

    # Tree structure
    'max_depth': randint(3, 12),
    'min_child_weight': randint(1, 10),
    'gamma': uniform(0, 5),

    # Learning parameters
    'learning_rate': uniform(0.01, 0.29),  # 0.01 to 0.3
    'subsample': uniform(0.6, 0.4),        # 0.6 to 1.0
    'colsample_bytree': uniform(0.6, 0.4), # 0.6 to 1.0
    'colsample_bylevel': uniform(0.6, 0.4),
    'colsample_bynode': uniform(0.6, 0.4),

    # Regularization
    'reg_alpha': uniform(0, 10),   # L1 regularization
    'reg_lambda': uniform(0, 10),  # L2 regularization

    'scale_pos_weight': uniform(1, 5),  # For class imbalance
}

# stratified k-fold
cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# RandomizedSearchCV
random_search = RandomizedSearchCV(
    estimator=xgb.XGBClassifier(
        objective='binary:logistic',
        random_state=42,
        n_jobs=-1,
        tree_method='hist',
        eval_metric='auc'
    ),
    param_distributions=param_distributions,
    n_iter=100,
    scoring='roc_auc',
    cv=cv_strategy,
    verbose=2,
    n_jobs=-1,
    random_state=42,
    return_train_score=True
)

sample_size = min(50000, len(X_train))
X_sample = X_train[:sample_size]
y_sample = y_train[:sample_size]

start_time = time.time()
random_search.fit(X_sample, y_sample)
elapsed_time = time.time() - start_time

print(f"\n✓ Randomized search completed in {elapsed_time/60:.2f} minutes")
print(f"\nBest parameters from Randomized Search:")
for param, value in random_search.best_params_.items():
    print(f"  {param:20s}: {value}")
print(f"\nBest cross-validation AUC: {random_search.best_score_:.4f}")
```

```python
print("\n[2/3] Training final model with best parameters...")
print("-" * 70)

# final model with the best hyperparameters from Randomized Search
best_params = random_search.best_params_
final_xgb_model = xgb.XGBClassifier(**best_params, objective='binary:logistic', random_state=42, n_jobs=-1, tree_method='hist')

eval_set = [(X_train, y_train), (X_test, y_test)]

start_time = time.time()
final_xgb_model.fit(X_train, y_train, eval_set=eval_set, verbose=False)
elapsed_time = time.time() - start_time

print(f"\n✓ Final model training completed in {elapsed_time/60:.2f} minutes")

print("\n[3/3] Evaluating final model on test set...")
print("-" * 70)

y_pred_final = final_xgb_model.predict(X_test)
y_pred_proba_final = final_xgb_model.predict_proba(X_test)[:, 1]

accuracy_final = accuracy_score(y_test, y_pred_final)
precision_final = precision_score(y_test, y_pred_final)
recall_final = recall_score(y_test, y_pred_final)
f1_final = f1_score(y_test, y_pred_final)
auc_final = roc_auc_score(y_test, y_pred_proba_final)

print("\nFinal Tuned Model Performance on Test Set:")
print(f"Accuracy: {accuracy_final:.4f}")
print(f"Precision: {precision_final:.4f}")
print(f"Recall: {recall_final:.4f}")
print(f"F1 Score: {f1_final:.4f}")
print(f"AUC: {auc_final:.4f}")
```

```python
import joblib

model_filename = 'final_xgb_model.joblib'

joblib.dump(final_xgb_model, model_filename)

print(f"Fine-tuned model saved to '{model_filename}'")
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve
import numpy as np
import seaborn as sns

# Plotting Learning Curves to check for overfitting
print("\nGenerating Learning Curve...")
print("-" * 70)

train_sizes, train_scores, test_scores = learning_curve(
    final_xgb_model, X_train, y_train, cv=5, scoring='roc_auc', n_jobs=-1,
    train_sizes=np.linspace(0.1, 1.0, 5)
)

train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)

plt.figure(figsize=(10, 6))
plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                 train_scores_mean + train_scores_std, alpha=0.1, color="r")
plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                 test_scores_mean + test_scores_std, alpha=0.1, color="g")
plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
         label="Training score (AUC)")
plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
         label="Cross-validation score (AUC)")
plt.title("Learning Curve")
plt.xlabel("Training examples")
plt.ylabel("AUC Score")
plt.legend(loc="best")
plt.grid(True)
plt.show()

# Feature Importance
print("\nGenerating Feature Importance Plot...")
print("-" * 70)

plt.figure(figsize=(12, 8))
xgb.plot_importance(final_xgb_model, max_num_features=20, importance_type='gain', title='Feature Importance (Gain)')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 8))
xgb.plot_importance(final_xgb_model, max_num_features=20, importance_type='weight', title='Feature Importance (Weight)')
plt.tight_layout()
plt.show()

# Confusion Matrix
print("\nGenerating Confusion Matrix...")
print("-" * 70)

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred_final)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Predicted Negative', 'Predicted Positive'],
            yticklabels=['Actual Negative', 'Actual Positive'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
